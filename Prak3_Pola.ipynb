```python
print("read text data from tsv file\n")

#Baca file tsv dengan lib np
traffic = np.genfromtxt("web_traffic.tsv",delimiter='\t')
print(traffic[:10])
print(traffic.shape)
```

```python
x = traffic[:,0]
y = traffic[:,1]
x = x[~np.isnan(y)]
y = y[~np.isnan(y)]
```

```python
#Melakukan plot dan menampilkan hasilnya
plt.scatter(x,y)
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
```

### Operasi Image

```python
#import library cv2
import cv2
```

```python
#Membaca file citra
five = cv2.imread("Citra -20210306/5.png")

print(five.shape) #mengecek dimensi citra
print(five.size) #mengecek ukuran citra

plt.imshow(five) #menampilkan citra
cv2.waitKey(0)
```

```python
# acces pixel of images per postion 
pixels = five[100,100]
print(pixels)
```

```python
#Kasus Babon
babon = cv2.imread("Citra -20210306/babon.jpg")

# convert gambar antara RGB/BGR and grayscale
babon_gray = cv2.cvtColor(babon, cv2.COLOR_BGR2GRAY)

plt.imshow(babon) #warna babon
```
